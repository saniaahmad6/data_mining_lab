```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
X = load_iris().data
X
```

```python
def eud(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def kmeans(x,k, max_iter = 100):
    centroids = x[:k]

    for _ in range(max_iter):
        clusters = [[] for _ in range(k)]
        
        for row in x:
            dis = [eud(row,centroid) for centroid in centroids]
            closest_idx = np.argmin(dis)
            clusters[closest_idx].append(row)

        old = centroids.copy()

        for i in range(k):
            if clusters[i]:
                centroids[i]=np.mean(clusters[i],axis=0)


        if np.all(old == centroids): 
            break

    return centroids, clusters
```

```python
def plot(centroids, clusters):
    colors = ['r', 'g', 'b', 'c', 'm']
    for ind, cluster in enumerate(clusters):
        cluster = np.array(cluster)
        plt.scatter(cluster[:,0],cluster[:,1],c=colors[ind],marker='o',label = f'Cluster {ind+1}')
    plt.scatter(centroids[:,0],centroids[:,1],c='yellow',marker='*',label = f'Centroids')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
k = 4
clusters, centroids = kmeans(X,k)
plot(clusters,centroids)
```
