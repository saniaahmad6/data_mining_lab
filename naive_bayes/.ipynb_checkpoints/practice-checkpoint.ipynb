```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table('./weather.txt')
df
```

```python
def pre_processing(df):
    y = df[df.columns[-1]]
    X = df.drop([df.columns[-1]],axis = 1)
    return X,y
X,y = pre_processing(df)
```
